```python
# Example daily temperatures in Celsius for 10 days
temperatures = [22.5, 23.0, 21.8, 24.2, 25.0, 23.5, 22.0, 24.5, 25.3, 23.8]

print(temperatures)
```

```python
# Number of days recorded
print("Number of days:", len(temperatures))

# Average temperature
avg_temp = sum(temperatures) / len(temperatures)
print("Average temperature:", avg_temp)

# Maximum and minimum
print("Maximum temperature:", max(temperatures))
print("Minimum temperature:", min(temperatures))
```

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({'temperature': temperatures})

print(df)

# Describe stats
print(df.describe())
```

```python
import matplotlib.pyplot as plt

plt.plot(df['temperature'], marker='o')
plt.title('Daily Temperatures')
plt.xlabel('Day')
plt.ylabel('Temperature (Â°C)')
plt.show()
```

df.to_csv('daily_temperatures.csv', index=False)
