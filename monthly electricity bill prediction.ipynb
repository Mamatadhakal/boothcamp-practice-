Predicting Monthly Electricity Bill

```python
#Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#define the dataset
#x is the average daily usage in hours which is our input feature
#y is the monthly bill in dollars whoch is our target findings/variables

x= np.array([[3], [4], [6], [7], [8], [10], [12]])
y= np.array([50, 60, 80, 95, 110, 130, 150])
```

```python
#creats the instance that is ready to fit the data
model = LinearRegression()
```

```python
#now we need to train the model on the given dataset
#now to find the best value that best fit line that minimizes errors between prediction and actual value
model.fit(x, y)
```

```python
# now we make prediction, so lets assume that the house uses electricity 5 hours a day
predicted_bill = model.predict (np.array([[5]]))
print("predicted monthly bill for 5 hours daily usage:",predicted_bill[0])
```

```python
# the slope which can show rate id incease in preice oer hour and the intercept is the baseline starting monthly bill
slope = model.coef_[0]
intercept = model.intercept_
print("slope (increase per additional hour):",slope)
print("intercept (base bill):", intercept)
```

```python
import matplotlib.pyplot as plt
```

```python
#plotting of the scatter plot
plt.scatter(x, y, color="green", label="data points")
#now predicting the entire range on x values for plotting the regression line

x_range = np.array([[x] for x in range(2, 14)]) #whats our range we want
y_pred = model.predict(x_range) #prediction of range of x

#plotting of regression line
plt.plot(x_range, y_pred, color='blue', label='regression line')

#question asked for uasage for 5 hours so,
# here s is size of the dot
plt.scatter(5, predicted_bill, color='black', 
    label="predicted bill", s=200, edgecolor='black')

#adding labels
plt.title('monthly electricity bill prediction')
plt.xlabel('average daily usage in hours')
plt.ylabel('monthly bill in dollars')
plt.legend()
plt.grid(True)
plt.show()
```
