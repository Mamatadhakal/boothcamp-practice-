This notebook generates random emojis and plots how many times each one appears. Have fun!

```python

import random
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline
```

## Generate Random Emojis

```python

# A list of emojis
emojis = ['üòÄ', 'üòÇ', 'üòç', 'ü§î', 'üòé', 'ü•≥', 'üò¥', 'üò±', 'ü§Ø', 'üò°']

# Generate 100 random emojis
random_emojis = [random.choice(emojis) for _ in range(100)]
print(' '.join(random_emojis))
```

## Count and Visualize Emoji Frequency

```python

# Count frequency
counts = Counter(random_emojis)

# Plot
plt.figure(figsize=(10,6))
plt.bar(counts.keys(), counts.values(), color='skyblue')
plt.title('Frequency of Random Emojis')
plt.xlabel('Emoji')
plt.ylabel('Count')
plt.show()
```
